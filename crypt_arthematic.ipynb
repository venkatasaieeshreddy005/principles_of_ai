```python
from itertools import permutations

def solve_cryptarithmetic(word1, word2, result_word):
    # Get all unique letters
    letters = list(set(word1 + word2 + result_word))

    # Cannot solve if more than 10 letters
    if len(letters) > 10:
        print("Too many letters! Cannot solve.")
        return

    # Try all possible digit assignments
    for perm in permutations(range(10), len(letters)):
        assignment = dict(zip(letters, perm))


        # First letter of a word cannot be 0
        if assignment[word1[0]] == 0 or assignment[word2[0]] == 0 or assignment[result_word[0]] == 0:
            continue

        # Convert words to numbers
        num1 = int("".join(str(assignment[ch]) for ch in word1))
        num2 = int("".join(str(assignment[ch]) for ch in word2))
        num_result = int("".join(str(assignment[ch]) for ch in result_word))

        # Check if sum is valid
        if num1 + num2 == num_result:
            print("Solution Found!")
            for letter in sorted(assignment):
                print(f"{letter} = {assignment[letter]}")
            print(f"{word1} + {word2} = {result_word}")
            print(f"{num1} + {num2} = {num_result}")
            return

    print("No solution found.")

# Example usage
solve_cryptarithmetic("SEND", "MORE", "MONEY")
```
