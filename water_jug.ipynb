```python
def print_board(board, n):
    for i in range(n):
        row = ""
        for j in range(n):
            if board[i] == j:
                row += "Q "
            else:
                row += ". "
        print(row)
    print()

def is_safe(board, row, col):
    for i in range(row):
        if board[i] == col or abs(board[i] - col) == row - i:
            return False
    return True

def solve_n_queens(board, row, n, solution_count):
    if row == n:
        solution_count[0] += 1
        print(f"Solution {solution_count[0]}:")
        print_board(board, n)
        return

    for col in range(n):
        if is_safe(board, row, col):
            board[row] = col
            solve_n_queens(board, row + 1, n, solution_count)

n = int(input("Enter number of queens: "))
board = [-1] * n
solution_count = [0]  # Using list to make it mutable

solve_n_queens(board, 0, n, solution_count)
print(f"Total solutions: {solution_count[0]}")
```
