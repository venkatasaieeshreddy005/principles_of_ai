```python
from heapq import heappush, heappop

def heuristic(state, target):
    return min(abs(state[0] - target), abs(state[1] - target))

def water_jug_a_star(capA, capB, target):
    start = (0, 0)
    open_list = []
    visited = set()

    heappush(open_list, (0, 0, start, []))

    while open_list:
        f, g, state, path = heappop(open_list)

        if state in visited:
            continue

        visited.add(state)
        path = path + [state]

        a, b = state

        if a == target or b == target:
            return path

        next_states = [
            (capA, b),  # Fill Jug A
            (a, capB),  # Fill Jug B
            (0, b),     # Empty Jug A
            (a, 0),     # Empty Jug B
            (max(0, a - (capB - b)), min(capB, b + a)),  # Pour A → B
            (min(capA, a + b), max(0, b - (capA - a)))   # Pour B → A
        ]

        for ns in next_states:
            if ns not in visited:
                g_new = g + 1
                h = heuristic(ns, target)
                heappush(open_list, (g_new + h, g_new, ns, path))

    return None


capA = 4
capB = 3
target = 2

solution = water_jug_a_star(capA, capB, target)

if solution:
    print("Steps:")
    for step in solution:
        print(step)
else:
    print("No solution found")
```
