```python
import heapq

goal = [[1,2,3],[4,5,6],[7,8,0]]

# Manhattan distance heuristic
def h(state):
    return sum(abs((val-1)//3 - i) + abs((val-1)%3 - j)
               for i,row in enumerate(state) for j,val in enumerate(row) if val)

# Get neighbor states by sliding empty tile
def neighbors(state):
    for i,row in enumerate(state):
        for j,val in enumerate(row):
            if val==0: x,y=i,j
    moves = [(-1,0),(1,0),(0,-1),(0,1)]
    result=[]
    for dx,dy in moves:
        nx,ny=x+dx,y+dy
        if 0<=nx<3 and 0<=ny<3:
            new=[r[:] for r in state]
            new[x][y],new[nx][ny]=new[nx][ny],new[x][y]
            result.append(new)
    return result

def a_star(start):
    pq = [(h(start),0,start,[])]
    visited = set()
    while pq:
        f,g,state,path = heapq.heappop(pq)
        tstate = tuple(tuple(r) for r in state)
        if tstate in visited: continue
        visited.add(tstate)
        if state==goal:
            for step in path+[state]:
                for r in step: print(r)
                print()
            print(f"Solved in {len(path)} moves!")
            return
        for n in neighbors(state):
            heapq.heappush(pq,(g+1+h(n),g+1,n,path+[state]))

# Example start
start=[[1,2,3],[4,0,6],[7,5,8]]
a_star(start)
```
